```python
import pandas as pd
import numpy as np
```

```python
tips = pd.read_csv('../datasets/tips.csv')

tips['tips_pct'] = tips['tip'] / tips['total_bill']
tips.head()
```

```python
tip_df_sorted_i = tips.sort_values(by = 'tips_pct', axis = 'index')
tip_df_sorted_i.head()
```

```python
def top_sorted_rows(df, column_name = 'tips_pct', n=5):
    
    sorted_df = df.sort_values(by = 'tips_pct', axis = 'index')
    
    return sorted_df[-n:]   
```

```python
tip_df_sorted_ii = top_sorted_rows(tips)
tip_df_sorted_ii
```

```python
smoker_df = tips.groupby('smoker')
smoker_df
```

```python
smoker_df.apply(top_sorted_rows, column_name = 'tips_pct', n = 3)
```

```python
# Call describe on groupby object
results = smoker_df['tips_pct'].describe()
results
```
