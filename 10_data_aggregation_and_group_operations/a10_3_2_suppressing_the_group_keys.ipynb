```python
import pandas as pd
import numpy as np
```

```python
tips = pd.read_csv('../datasets/tips.csv')

tips['tips_pct'] = tips['tip'] / tips['total_bill']
tips.head()
```

```python
grp_i = tips.groupby('smoker', group_keys = False)
grp_i
```

```python
def top_sorted_rows(df, column_name = 'tips_pct', n=5):
    
    sorted_df = df.sort_values(by = 'tips_pct', axis = 'index')
    
    return sorted_df[-n:]
```

```python
result_i = grp_i.apply(top_sorted_rows)
result_i
```
