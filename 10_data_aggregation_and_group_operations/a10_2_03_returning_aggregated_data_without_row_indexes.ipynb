#### Returning Aggregated Data Without Row Indexes

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../datasets/tips.csv')
df.head()
```

```python
grp_i = df.groupby(by = ['day', 'smoker'], as_index= False)
grp_i
```

```python
for grp_name, grp_data in grp_i:
    print(grp_name)
    print(grp_data)
    print('=='*20)
```

```python
grp_i.mean()
```
